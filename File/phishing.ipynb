```python
X = data.drop(columns=['Result', 'having_IP_Address'])
y = data['Result']

from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform y to convert -1 to 0 and 1 stays 1
y = label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
import os
import joblib  # You can use joblib for saving models

# Correct usage of train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=456)

class ModelTrainer:
    def __init__(self, trained_model_path: str):
        # Store path for saving trained models
        self.trained_model_path = trained_model_path
        
        # Initialize models
        self.models = {
            "GaussianNB": GaussianNB(),
            "XGBClassifier": XGBClassifier(objective='binary:logistic'),
            "LogisticRegression": LogisticRegression()
        }
        
        # Default parameter grid (example for XGBClassifier)
        self.model_param_grid = {
            "XGBClassifier": {
                'max_depth': [3, 5, 7],
                'learning_rate': [0.01, 0.1],
                'n_estimators': [100, 200]
            },
            "LogisticRegression": {
                'C': [0.1, 1.0, 10.0],
                'penalty': ['l2', 'l1']
            },
            "GaussianNB": {}  # GaussianNB has no hyperparameters for GridSearch
        }

    def evaluate_models(self, X_train, X_test, y_train, y_test):
        report = {}

        for model_name, model in self.models.items():
            model.fit(X_train, y_train)  # Train model

            y_train_pred = model.predict(X_train)
            y_test_pred = model.predict(X_test)

            train_model_score = accuracy_score(y_train, y_train_pred)
            test_model_score = accuracy_score(y_test, y_test_pred)

            report[model_name] = test_model_score

        return report

    def get_best_model(self, X_train, y_train, X_test, y_test):
        model_report = self.evaluate_models(X_train, X_test, y_train, y_test)
        
        print(model_report)

        # Get the best model based on test score
        best_model_name = max(model_report, key=model_report.get)
        best_model_object = self.models[best_model_name]
        best_model_score = model_report[best_model_name]

        return best_model_name, best_model_object, best_model_score

    def finetune_best_model(self, best_model_object, best_model_name, X_train, y_train):
        param_grid = self.model_param_grid.get(best_model_name, {})
        
        if param_grid:  # Only perform grid search if parameters are provided
            grid_search = GridSearchCV(best_model_object, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
            grid_search.fit(X_train, y_train)
            best_params = grid_search.best_params_

            print(f"Best params for {best_model_name}: {best_params}")

            # Update model with the best parameters found
            best_model_object.set_params(**best_params)

        return best_model_object

    def initiate_model_trainer(self, X_train, y_train, X_test, y_test):
        # Get the best model and its score
        best_model_name, best_model_object, best_model_score = self.get_best_model(X_train, y_train, X_test, y_test)

        # Fine-tune the best model
        best_model_object = self.finetune_best_model(best_model_object, best_model_name, X_train, y_train)

        # Train the best model on the full training data
        best_model_object.fit(X_train, y_train)

        # Evaluate the model on test data
        y_pred = best_model_object.predict(X_test)
        final_model_score = accuracy_score(y_test, y_pred)

        print(f"Best model: {best_model_name} with accuracy: {final_model_score}")

        if final_model_score < 0.6:
            raise Exception(f"No best model found with an accuracy greater than the threshold 0.6")

        # Ensure the directory exists for saving the model (if a directory is part of the path)
        model_directory = os.path.dirname(self.trained_model_path)
        if model_directory:  # If the path includes a directory
            os.makedirs(model_directory, exist_ok=True)
        
        # Save the trained model to disk
        joblib.dump(best_model_object, self.trained_model_path)  # Save model with joblib

        return final_model_score

# Instantiate the ModelTrainer with a path to save the model
model_trainer = ModelTrainer("trained_model.pkl")

# Call initiate_model_trainer to start the training process
model_score = model_trainer.initiate_model_trainer(X_train, y_train, X_test, y_test)
print(f"Final Model Score: {model_score}")
```
